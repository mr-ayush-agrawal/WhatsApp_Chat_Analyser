## Pre Processing the Chat 

```python
import re
import pandas as pd
```

```python
ChatFile = open('GroupChat.txt', encoding='utf-8')
Chats = ChatFile.read()
print(Chats)
```

**Splitting the Chat as date/time , Sender, Message**

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
#  This is the re pattern till the date, before starting the Sender's Name
```

```python
Messages = re.split(pattern,Chats)[1:]
Messages
```

```python
TimeStamp = re.findall(pattern, Chats)
TimeStamp
```

#### Making DataFrame

```python
Data = pd.DataFrame({'Stamp':TimeStamp, "Message": Messages})
# Convertinng the Date time to pandas DateTime
Data['Stamp'] = pd.to_datetime(Data['Stamp'], format='%m/%d/%y, %H:%M - ')
Data.shape
```

### Splitting the messages

```python
Senders = []
Messages = []

for msg in Data['Message'] :
    ent = re.split('([\w\W]+?):\s', msg)
    if ent[1:] :            # If having any sender
        Senders.append(ent[1])
        Messages.append(ent[2].lower())
    else :
        Senders.append("Notification")
        Messages.append(ent[0].lower())

# type(ent[0])

Data['Message'] = Messages
Data['Sender']  = Senders

Data.head(10)

# If this would Run Multiple times then all the senders would be Updated to Notification as this wont find any user in the list
```

### Extracting the Date, Time, Month year of the message

Since it is in DateTime format we Can use the attributes for the task

```python
Data['Year'] = Data['Stamp'].dt.year
Data['Month'] = Data['Stamp'].dt.month_name()
Data['Day'] = Data['Stamp'].dt.day
Data['Hour'] = Data['Stamp'].dt.hour
Data['Minute'] = Data['Stamp'].dt.minute
Data.drop('Stamp', axis=1, inplace=True)
Data.head()
```
